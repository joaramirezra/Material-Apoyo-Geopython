```python
# Caso de uso 
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

img=mpimg.imread('mapa.jpg')

def poligono_sobre_seccion_Delgada():

  coor_x = []
  coor_y = []

  puntos = int(input("ingrese la cantidad de aristas del poligono : "))
  for punto in range (puntos):
    x,y = input("ingrese las coordenada "+str(punto) + " del poligono separada por espacio : ").split(" ")
    coor_x.append(int(x))
    coor_y.append(int(y))

  plt.figure()
  plt.imshow(img)
  plt.plot(coor_x,coor_y,'r*-')
  plt.show()
  
```

```python
poligono_sobre_seccion_Delgada()
```
