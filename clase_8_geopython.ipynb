```python
import mplstereonet as mpl
import matplotlib.pyplot as plt
```

```python
ang_plano,dir_plano = 35,150 
ang_falla,dir_falla = 18,39

ang_z,dir_z = mpl.plane_intersection(ang_plano,dir_plano,ang_falla,dir_falla)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection='stereonet')

ax.plane(ang_plano,dir_plano,c='c',label = 'plano')
ax.plane(ang_falla,dir_falla,'r',label = 'falla',alpha=0.3)
ax.line(ang_z,dir_z,label = 'interseccion')
plt.title('Red esterografica ',pad = 30, fontsize = 15)
plt.legend()
plt.grid(1)
plt.show()

```

```python
def graficar_red(*fallas):
  
  fig = plt.figure(figsize=(8,8))
  ax = fig.add_subplot(111,projection='stereonet')

  for falla in fallas :
    ang_plano,dir_plano,label_falla= falla
    ax.plane(ang_plano,dir_plano,label = label_falla)

  
  plt.title('Red esterografica ',pad = 30, fontsize = 15)
  plt.grid(1)
  plt.legend()
  plt.show()
```

```python
graficar_red([5,10,'plano falla'],[52,10,'plano corte'])
```

```python
graficar_red([15,15,'falla1'],[18,130,'falla2'],[18,4,'falla3'],[93,150,'falla 4'])
```

```python
def graficar_estratificacion_fallada(falla,plano):
  ang_fall, dir_falla = falla
  ang_plano, dir_plano = plano
  plunge,bearing = mplstereonet.plane_intersection(ang_fall, dir_falla,
                                                   ang_plano, dir_plano)
                                                   
  fig = plt.figure(figsize=(8,8))
  ax = fig.add_subplot(111,projection='stereonet')
  ax.plane(ang_fall, dir_falla, c='b',label = 'falla')
  ax.plane(ang_plano, dir_plano,c='r',label = 'plano')
  ax.line(plunge,bearing, 'k*', label = ' interseccion')
  plt.legend()
  plt.grid(1)
  plt.show()
  print('coordenadas de intersecion ',plunge,bearing)
```
